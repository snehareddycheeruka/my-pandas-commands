```python
#clean experience years by converting text values to numeric and remove negative values
def convert_experience(exp):
    try:
        years = int(exp)
        return years if years >= 0 else pd.NA
    except:
        return pd.NA
df['experience_years'] = df['experience_years'].apply(convert_experience)
df['experience_years']
```

```python
#convert is_active values to yes/no true/false boolean format 
def convert_is_active(val):
    val = str(val).strip().lower()
    if val in ['yes', 'true', '1']:
        return True
    elif val in ['no', 'false', '0']:
        return False
    else:
        return pd.NA
df['is_active'] = df['is_active'].apply(convert_is_active)
df['is_active']
```
